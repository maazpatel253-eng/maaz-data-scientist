```python
# K-Means Clustering
#Project: Customer Segmentation
#Objective

# Group customers based on income and spending score.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = {
    "Income": [15,16,17,18,50,52,54,55,90,92,95,98],
    "Spending": [39,40,42,45,60,62,65,68,80,82,85,88]
}

df = pd.DataFrame(data)

kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(df[["Income","Spending"]])

plt.scatter(df["Income"], df["Spending"], c=df["Cluster"])
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")
plt.show()

print(df)
```

```python
#explanation
#K-Means creates customer groups.
#Customers in the same cluster have similar behavior.
#Businesses use this for targeted marketing.
```

```python
#Decision Tree
#Project: Student Pass/Fail Prediction
#Objective

#Predict whether a student will pass based on study hours.
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Result": [0,0,0,0,1,1,1,1]
}

df = pd.DataFrame(data)

X = df[["Hours"]]
y = df["Result"]

model = DecisionTreeClassifier()
model.fit(X,y)

prediction = model.predict([[5]])
print("Prediction:", prediction)
```

```python
#Explanation
#Decision Tree learns rules.
#Example:
#Hours < 5 → Fail
#Hours ≥ 5 → Pass
```

```python
#ime Series Forecasting
#Project: Sales Forecasting
#Objective

#Predict future sales.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

sales = [100,120,130,150,170,180,200]

df = pd.DataFrame({
    "Month": range(1,8),
    "Sales": sales
})

df["Moving_Avg"] = df["Sales"].rolling(3).mean()

plt.plot(df["Month"], df["Sales"], marker="o")
plt.plot(df["Month"], df["Moving_Avg"], marker="o")
plt.legend(["Sales","Forecast"])
plt.show()

next_sales = sum(sales[-3:])/3
print("Predicted Next Month Sales:", next_sales)
```

```python
#Explanation
#Uses Moving Average forecasting.
#Predicts future sales using past trends.
```

```python
#Weather Detection
#Project: Rain Prediction
#Objective

#Predict whether it will rain.
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data = {
    "Temp":[35,30,25,20,32,22,18],
    "Humidity":[40,50,80,90,45,85,95],
    "Rain":[0,0,1,1,0,1,1]
}

df = pd.DataFrame(data)

X = df[["Temp","Humidity"]]
y = df["Rain"]

model = LogisticRegression()
model.fit(X,y)

prediction = model.predict([[24,88]])
print("Rain Prediction:", prediction)
```

```python
#Explanation
#Uses temperature and humidity.
#Predicts Rain (1) or No Rain (0)
```

```python
#lassification
#Project: Iris Flower Classification
#Objective

#Identify flower species.
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    test_size=0.2,
    random_state=42
)

model = RandomForestClassifier()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
```

```python
#Explanation
#Multi-class classification.
#Predicts flower type:
#Setosa
#Versicolor
#Virginica
```

```python
#Naive Bayes Filtering
#Project: Spam Email Detection
#Objective

#Detect spam messages.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

texts = [
    "win money now",
    "free lottery ticket",
    "hello friend",
    "how are you"
]

labels = [1,1,0,0]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = MultinomialNB()
model.fit(X, labels)

test = vectorizer.transform(["free money"])
prediction = model.predict(test)

print("Spam Prediction:", prediction)
```

```python
#Explanation
#Converts text into numbers.
#Naive Bayes calculates probabilities.
#Predicts:
#1 = Spam
#0 = Not Spam
```
